```python
train_data, test_data, val_data = tdata.random_split(data, (33600, 6300, 2100))
```

```python
batch_size = 10

train_dataloader = tdata.DataLoader(train_data, batch_size=batch_size, shuffle=True, num_workers=4)
test_dataloader = tdata.DataLoader(test_data, batch_size=batch_size, shuffle=True, num_workers=4)
val_dataloader = tdata.DataLoader(val_data, batch_size=batch_size, shuffle=True, num_workers=4)

dataloaders = {'train':train_dataloader, 'test':test_dataloader, 'val':val_dataloader}
```

```python
model = tmodels.MobileNetV2(num_classes=10)
model.features[0][0] = nn.Conv2d(1, 32, kernel_size=(3, 3), stride=(2, 2), padding=(1, 1), bias=False)
```

```python
ce_loss = nn.CrossEntropyLoss()
calculate_loss = lambda pred, true: ce_loss(pred, true)
```

```python
opt = torch.optim.Adam(model.parameters(), lr=1e-3)
```

```python
epoch_i = 0

losses = {'train':[], 'test':[]}
```

```python
n_epochs = 100
```

```python
while epoch_i <= n_epochs:
    
    for mode in ['train', 'test']:
        
        if mode=='train':
            model.train()
        else:
            model.eval()
            
        for feats, target in dataloaders[mode]:
            pred = model(feats)
            loss = calculate_loss(pred, target)

            losses[mode].append(loss.item())
            
            if mode=='train':
                loss.backward()
                opt.step()
                opt.zero_grad()
                
    epoch_i += 1
    
    if epoch_i % 10 == 0:
        clear_output(True)
        visualize_loss(losses)
```
